```python
nfcount = 0
notfraud = df["Class"]
for i in range(len(notfraud)):
    if notfraud[i] == 0:
        nfcount = nfcount+1
nfcount
per_nf = (nfcount/len(notfraud))*100
print(per_nf)
```

```python
fcount = 0
fraud = df["Class"]
for i in range(len(fraud)):
    if fraud[i] == 1:
        fcount = fcount+1
nfcount
per_f = (fcount/len(fraud))*100
print(per_f)
```
